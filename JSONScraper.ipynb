```python
newlist = []
for names in fnames:
    newlist.append(names.split("/"))
```

```python
list3 = []
for i in range(len(newlist)):
    list3.append(newlist[i][-1])
```

```python
list4 = []
for names in list3:
    list4.append(names.split("_"))
```

```python
list5 = []
for i in range(len(list4)):
    list5.append(list4[i][2])
```

```python
uniqueList = []
duplicateList = []

for i in list5:
    if i not in uniqueList:
        uniqueList.append(i)
    elif i not in duplicateList:
        duplicateList.append(i)
```

```python
investigate = duplicateList
dataInd = []
for i in range(len(fnames)):
    for k in investigate:
        if k in fnames[i]:
            dataInd.append(i)
            
```

```python
data = [jsonload(fname) for fname in fnames]

for i in range(len(data)):
    if data[i] == None:
        data.remove(data[i])
```

```python
for j in range(len(dataInd)):
    for i in range(len(data[dataInd[j]]['tests'])):
        if 'failed' in data[dataInd[j]]['tests'][i]['outcome']:
            print("chipNum:", list5[dataInd[j]])
#             print("test index:",i)
#             print("test name:", data[dataInd[0]]['tests'][i]['nodeid'])
#             print("failure cause:", data[dataInd[0]]['tests'][i]['call']['traceback'][0]['message'])
#             #print("failure location:", data[dataInd[0]]['tests'][i]['call']['crash']['message'])
```

```python
headers = ["chipNum", "dataInd", "testInd", "failureMode", "codeLine", "testName",  "failureCause"]
workbook = openpyxl.Workbook()
worksheet = workbook.active
worksheet.append(headers)
for j in range(len(dataInd)):
    for i in range(len(data[dataInd[j]]['tests'])):
        if 'failed' in data[dataInd[j]]['tests'][i]['outcome']:
            chipNum = list5[dataInd[j]]
            dataIndex = dataInd[j]
            testInd = i
            testName = data[dataInd[j]]['tests'][i]['nodeid']
            if data[dataInd[j]]['tests'][i]['call']['traceback'][0]['message'] != '':
                failureMode = data[dataInd[j]]['tests'][i]['call']['traceback'][0]['message']
            else:
                failureMode = data[dataInd[j]]['tests'][i]['call']['crash']['message']
            failureCause = data[dataInd[j]]['tests'][i]['call']['crash']['message']
            codeLine = data[dataInd[j]]['tests'][i]["call"]["crash"]["lineno"]
            X = [chipNum, dataIndex, testInd, failureMode, codeLine, testName, failureCause]
            X = [str(x) if isinstance(x, np.generic) else x for x in X]
            worksheet.append(X)
workbook.save('Failed_chip_retested.xlsx')
```
